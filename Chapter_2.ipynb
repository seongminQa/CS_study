# 2장 **데이터**

- 컴퓨터가 이해하는 정보 단위를 이해한다.
- 0과 1로 다양한 숫자를 표현하는 방법을 이해한다.
- 0과 1로 다양한 문자를 표현하는 방법을 이해한다.

## 2-1  **0과 1로 숫자를 표현하는 방법**

컴퓨터는 0과 1로 모든 정보를 표현하고, 0과 1로 표현된 정보만을 이해할 수 있다.
<br><br>

### **정보 단위**

- 비트(bit) : 0과 1을 나타내는 가장 작은 정보 단위
  - 1비트의 정보 단위 : 2가지
  - 2비트의 정보 단위 : 4가지
  - 3비트의 정보단위 : 8가지

<br>

웹 브라우저, 워드 프로세서, 포토샵 등 우리가 실행하는 모든 프로그램은 수십만, 수백만 개 이상의 0과 1로 이루어져 있다.
<br><br>

- **1바이트(1byte)** : 8비트(8bit)
- **1킬로바이트(1kB)** : 1,000바이트(1,000byte)
- **1메가바이트(1MB)** : 1,000킬로바이트(1,000kB)
- **1기가바이트(1GB)** : 1,000메가바이트(1,000MB)
- **1테라바이트(1TB)** : 1,000기가바이트(1,000GB)
<br>
  cf) 1,024개씩 묶어 표현한 단위는 KiB, MiB, GiB, TiB이다.

<br>

* 워드(word) : CPU가 한 번에 처리할 수 있는 데이터 크기를 의미 <br>
  cf) 현대 컴퓨터의 워드 크기는 대부분 32비트 or 64비트이다.<br>
  가령 인텔의 x86 CPU는 32비트 워드 CPU, <br>
  x64 CPU는 64비트 워드 CPU이다.

<br><br>

## **이진법**

수학에서 0과 1만으로 모든 숫자를 표현하는 방법을 **이진법(binary)**라고 한다. <br>
이진법으로 표현한 수를 **이진수**, 0과 1밖에 모르는 컴퓨터에 어떤 숫자를 알려 주려면 십진수가 아닌 이진수로 알려주어야 한다. 코드 상에서 이진수를 표기할 때 이진수 앞에 **0b**를 붙인다.
<br><br>

## **이진수의 음수 표현**

- 2의 보수(two's complement) : 음수를 표현하는 방법 중 가낭 널리 사용되는 방법
  - 양수인지 음수인지 구분하기 위한 **플래그(flag)**가 있다.

<br><br>

## **십육진법**

이진법을 이용하면 컴퓨터가 이해하는 숫자 정보를 직접적으로 표현할 수 있다.<br>
하지만 숫자의 길이가 너무 길어진다는 단점이 있다. 그래서 데이터를 표현 할 때 이진법 이외에 **십육진법**도 자주 사용한다.<br>
코드 상에 십육진법을 표기할 때 앞에 **0x**를 붙인다.
<br><br>

## **십육진수를 이진수로 변환하기**

십육진수 한 글자를 4비트의 이진수로 간주한다.
<br><br>

## **이진수를 십육진수로 변환하기**

이진수 숫자를 네 개씩 끊고, 끊어준 네 개의 숫자를 하나의 십육진수로 변환한 뒤 그대로 이어 붙인다.
<br><br>

cf) 이진수를 십진수로 변환할 때는 이렇게 간단하지 않기에 이진수를 십육진수로 묶어 표현하는 것이다. <br>
십육진수는 프로그래밍할 때 이진수와 더불어 자주 사용되므로 기억해 두는 것이 좋다. 하드웨어와 밀접하게 맞닿아 있는 개발 분야에서는 코드에 십육진수를 직접 쓰는 경우도 더러 있기 때문이다.


## 2-2 **0과 1로 문자를 표현하는 방법**

### **문자 집합과 인코딩**

- **문자 집합(character set)** : 컴퓨터가 인식하고 표현할 수 있는 문자의 모음<br>
  (문자 집합에 속해 있지 않은 문자는 이해할 수 없다)
- **문자 인코딩(character encoding)** : 문자를 0과 1로 변환하는 과정 (컴퓨터가 이해할 수 있게 변환)
- **문자 디코딩(character decoding)** : 0과 1로 이루어진 문자 코드를 사람이 이해할 수 있는 문자로 변환하는 과정

<br><br>

### **아스키(ASCII) 코드**

https://stepbystep1.tistory.com/10 <br>

초창기 문자 집합 중 하나로, 영어 알파벳과 아라비아 숫자, 그리고 일부 특수 문자를 포함한다.정보의 가짓수는 2의 7승개로, 총 128개의 문자를 표현할 수 있다.<br><br>
0부터 127까지 총 128개의 숫자 중 하나의 고유한 수에 **일대일 대응**된다. 아스키 문자에 대응된 고유한 수를 **아스키 코드**라고 한다.
<br><br>
cf) 실제로 하나의 아스키 문자를 나타내기 위해 8비트(1바이트)를 사용한다. 하지만 8비트 중 1비트는 **패리티 비트(parity bit)**라고 불리는, **오류 검출을 위해 사용되는 비트**이기 때문에 실질적으로 문자 표현을 위해 사용되는 비트는 7비트이다.
<br><br>

아스키 코드는 매우 간단하게 인코딩된다는 장점이 있지만, 한글뿐만 아니라 아스키 문자 집합 외의 문자, 특수문자도 표현할 수 없는 단점도 있다.
<br><br>

- **EUC-KR** : 한글 인코딩 방식, 한글 완성형 인코딩 방식<br>
  (모든 한글을 표현할 수 없다는 한계가 있음)
- **유니코드** : EUC-KR보다 훨씬 다양한 한글을 포함하며 대부분 나라의 문자, 특수문자, 화살표나 이모티콘까지도 코드로 표현할 수 있는 통일된 문자 집합이다.<br>
  현대 문자를 표현할 때 가장 많이 사용되는 표준 문자 집합이며, 문자 인코딩 세계에서 매우 중요한 역할을 맡고 있다.
  - 아스키 코드나 EUC-KR은 글자에 부여된 값을 그대로 인코딩 값으로 삼는다.<br>
  유니코드는 글자에 부여된 값 자체를 인코딩된 값으로 삼지 않고 이 값을 다양한 방법으로 인코딩한다. 이러한 방법에는 UTF-8, UTF-16, UTF-32등이 있다.<br>
  cf) UTF = Unicode Transformation Format
